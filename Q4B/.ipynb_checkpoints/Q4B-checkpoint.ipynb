```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
from qutip import *
from scipy import *
```

```python
from h5py import File
    
def read_h5( iter: int , ham : int, ovFlag = False, level = 0, krylovs = []):
        x =  list(File(f"result.h5")[f"{iter}.w.{ham}.riz.{2-ovFlag}.0.0.0"])
        l = int(np.sqrt(len(x[0])))
        x =  np.reshape(x,(l,l))
        return np.array([[ x[i][j] for i in krylovs ] for j in krylovs])
H = {}
krylovs = range(64)
for i,n in enumerate(['boson','dicke','external','ising','z-1','z-2','z-3','z-4','d-1','d-2','d-3','d-4','a']):
    print(i,n)
    if n in H:
        H[n] += read_h5(1, i+1,krylovs = krylovs)
    else:
        H[n] =  read_h5(1, i+1,krylovs= krylovs)
o = read_h5( 1, 1, True,krylovs = krylovs)
ea,ev = linalg.eig(o)
'condition number', np.sort(ea)[-1]/np.sort(ea)[0]
```

```python
tlist  = np.linspace(0.0, 50, 1001)
A     = 0.1
psi0   = Gs
```

```python
import numpy as np
from matplotlib import pyplot
import qutip

   
def p(omega):
    T      = 2*np.pi / omega
    args = {'w': omega}
    H = [H0, [A * H1, lambda t, w: np.sin(w * t)]]
    
    # find the floquet modes for the time-dependent hamiltonian
    floquetbasis = qutip.FloquetBasis(H, T, args, precompute=tlist)
    
    # decompose the inital state in the floquet modes
    f_coeff = floquetbasis.to_floquet_basis(psi0)
    
    # calculate the wavefunctions using the from the floquet modes coefficients
    p_ex = [ [ qutip.expect(obs,floquetbasis.from_floquet_basis(f_coeff, t)) for t in tlist ] for obs in Obs]
    return tlist,p_ex

pyplot.plot(tlist, np.real(p(10)[1][0]) )#,np.real(p(1)[1][0]),tlist,np.real(p(10)[1][0]))


pyplot.xlabel('Time')
pyplot.ylabel('Photon Count')
pyplot.legend((f'$\lambda$ = {round(2*np.pi/10,2)}', ))
pyplot.show()
```
